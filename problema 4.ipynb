```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

def calcular_histograma_grises(imagen):
    # Cargar la imagen en color
    bgr_image = cv2.imread(imagen)

    if bgr_image is None:
        raise ValueError(f"No se pudo leer la imagen en la ruta: {imagen}")

    # Obtener los canales de color
    blue_channel = bgr_image[:, :, 0]
    green_channel = bgr_image[:, :, 1]
    red_channel = bgr_image[:, :, 2]

    # Calcular el promedio aritmético para escala de grises
    gray_img = np.mean(bgr_image, axis=-1)

    # Calcular los histogramas para cada canal de color y escala de grises
    hist_red, _ = np.histogram(red_channel, bins=256, range=(0, 256))
    hist_green, _ = np.histogram(green_channel, bins=256, range=(0, 256))
    hist_blue, _ = np.histogram(blue_channel, bins=256, range=(0, 256))
    hist_gray, _ = np.histogram(gray_img, bins=256, range=(0, 256))

    # Calcular la media de la distribución
    media_red = np.mean(red_channel)
    media_green = np.mean(green_channel)
    media_blue = np.mean(blue_channel)
    media_gray = np.mean(gray_img)

    # Mostrar los histogramas con la línea vertical que muestra la media
    plt.figure(figsize=(10, 6))

    plt.subplot(221)
    plt.plot(hist_red, color='red')
    plt.axvline(x=media_red, color='black', linestyle='dashed', linewidth=1)
    plt.title('Histograma Canal Rojo')
    plt.xlabel('Valor de píxel')
    plt.ylabel('Frecuencia')

    plt.subplot(222)
    plt.plot(hist_green, color='green')
    plt.axvline(x=media_green, color='black', linestyle='dashed', linewidth=1)
    plt.title('Histograma Canal Verde')
    plt.xlabel('Valor de píxel')
    plt.ylabel('Frecuencia')

    plt.subplot(223)
    plt.plot(hist_blue, color='blue')
    plt.axvline(x=media_blue, color='black', linestyle='dashed', linewidth=1)
    plt.title('Histograma Canal Azul')
    plt.xlabel('Valor de píxel')
    plt.ylabel('Frecuencia')

    plt.subplot(224)
    plt.plot(hist_gray, color='gray')
    plt.axvline(x=media_gray, color='black', linestyle='dashed', linewidth=1)
    plt.title('Histograma Escala de Grises')
    plt.xlabel('Valor de píxel')
    plt.ylabel('Frecuencia')

    plt.tight_layout()
    plt.show()
```

```python
# Ejemplo de uso:
imagen = "montefuji.jpg"  # Reemplaza con la ruta de tu imagen en color

# Mostrar los histogramas con la media de la distribución
calcular_histograma_grises(imagen)
```
